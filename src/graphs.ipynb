```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Load CSV
csv_path = r"D:\Akash\Study\FinalYear\VehicleIntrusionAlarmSystem\models\runs\kaggle\working\runs\detect\vehicle_detection\Teacher\results.csv"
df = pd.read_csv(csv_path)

# Professional style
sns.set_theme(style="ticks")
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 13,
    "axes.titlesize": 16,
    "axes.labelsize": 14,
    "legend.fontsize": 12
})

save_dir = "thesis_plots"
os.makedirs(save_dir, exist_ok=True)

# Metric names with clean titles
metric_titles = {
    "train/box_loss": "Training Box Loss",
    "train/cls_loss": "Training Classification Loss",
    "train/dfl_loss": "Training Distribution Focal Loss",
    "val/box_loss": "Validation Box Loss",
    "val/cls_loss": "Validation Classification Loss",
    "val/dfl_loss": "Validation Distribution Focal Loss",
    "metrics/precision(B)": "Precision",
    "metrics/recall(B)": "Recall",
    "metrics/mAP50(B)": "mAP@50",
    "metrics/mAP50-95(B)": "mAP@50–95"
}

for metric, title in metric_titles.items():
    
    plt.figure(figsize=(8,5))
    
    sns.lineplot(
        x=df["epoch"],
        y=df[metric],
        linewidth=2.8,
        marker="o"
    )

    plt.title(title, pad=12)
    plt.xlabel("Epoch")
    plt.ylabel("Value")
    
    sns.despine()
    
    plt.tight_layout()

    filename = metric.replace("/", "_")
    plt.savefig(f"{save_dir}/{filename}.png", dpi=400, bbox_inches="tight")
    
    plt.show()
```
